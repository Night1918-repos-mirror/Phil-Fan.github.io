## 题目一
一个有9面的骰子，连续投掷哪些可能，对应的概率大小?
- 用随机数生成的方法得到两次投掷的结果，并计算其和的所有可能和概率
- (dataframe)展示将以上结果整理为表格
- 将以上结果绘制为图片,两次投掷之和与概率用x和y轴代表

```python
# 01 连续投掷哪些可能,概率大小
import itertools
import numpy as np
import collections

dice = np.arange(1,10)
y = list(itertools.product(dice,dice))
sum_of_dice = collections.Counter([sum(x) for x in y])
sorted_sum_of_dice = list(sorted(sum_of_dice.items(),key=lambda x:x[0]))
output = [i[1]/len(y) for i in sorted_sum_of_dice]
print('共有',len(y),"种等可能情况")

# 02 用随机数生成的方法得到两次投掷的结果，并计算其和的所有可能和概率

sample = 2000
random_dice = np.random.randint(1,10,size=(2,sample))
sum_of_random_dice = collections.Counter([sum(x) for x in random_dice.T])
sorted_sum_of_random_dice = list(sorted(sum_of_random_dice.items(),key=lambda x:x[0]))
output_random = [i[1]/sample for i in sorted_sum_of_random_dice]
print("两次点数之和共有",len(sorted_sum_of_random_dice),"种可能性")

# 03 (dataframe)展示将以上结果整理为表格

import pandas as pd
df = pd.DataFrame(zip(list(range(2,19)),output,output_random),columns=['sum','理想概率','实验概率'])
print(df)

# # 04 将以上结果绘制为图片,两次投掷之和与概率用x和y轴代表
import seaborn as sns
import matplotlib.pyplot as plt

f, axs = plt.subplots(1, 2, figsize=(12, 6))
sns.histplot([sum(x) for x in y], stat='probability', ax=axs[0], bins=range(2, 19),discrete=True)
axs[0].set_title('ideal')
sns.histplot(random_dice[0]+random_dice[1] ,stat='probability', ax=axs[1], bins=range(2, 19),discrete=True)
axs[1].set_title('experiment')
plt.tight_layout()  # 调整子图之间的间距
```

## 题目2
连续投掷两次骰子，一直点数总和为9，第一次为6的可能性有多大

```python
import collections
import numpy as np
import itertools
sample = 10000
x = np.random.randint(1, 7, size=(2, sample))
calc = collections.Counter(x[0]+x[1]==9)
calc_both = collections.Counter((x[0]==6) & (x[1]==3))
print("理想概率为",1/4)
print("实验概率为",calc_both[True]/calc[True])
```
